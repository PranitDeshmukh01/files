```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import warnings 
warnings.filterwarnings("ignore")
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
Data=load_iris()
```

```python
df=pd.DataFrame(Data.data,columns=Data.feature_names)
```

```python
df
```

```python
df["Target"]=Data.target
```

```python
df
```

```python
x=df.iloc[:,0:4]
y=df.iloc[:,4]
```

```python
x
```

```python
y
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,train_size=0.8,random_state=20)
```

```python
Logistic_regression_model=LogisticRegression()
```

```python
Logistic_regression_model.fit(x_tr,y_tr)
```

```python
Logistic_regression_model.score(x_tr,y_tr)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
y_pred=Logistic_regression_model.predict([[5.1,3.8,1.4,0.4]])
```

```python
y_pred=Logistic_regression_model.predict(x_te)
```

```python
accuracy_score(y_pred,y_te)
```

```python
confusion_matrix(y_pred,y_te)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(y_pred,y_te),annot=True,cmap=["red","black","yellow"])
```
